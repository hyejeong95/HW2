선으로 그래프를 이어주고 싶으면 geom_point가 아닌 geom_line으로 해준다. 하지만 geom_line을 이용하면 모든 점들을 다 이어주기 떄문에 geom_smooth를 이용해서 그래주면 전반적인 그래프 형태가 나온다. 

```r
ggplot(data=mpg)+
geom_line(mapping=aes(x=displ,y=hwy))
```

```r
ggplot(data=mpg)+
geom_smooth(mapping=aes(x=displ,y=hwy))+
geom_line(mapping=aes(x=displ,y=hwy))
```
